**Pistas Iniciales:**


1.   Tipo de crimen: 'murder'.
2.   Fecha del crimen: 15 de enero de 2018.
3.   Ubicación del crimen: SQL City.

**Paso 1: Investigar el Informe de la Escena del Crimen:**

Este resultado indica que hubo un asesinato en SQL City el 15 de enero de 2018, y que hubo dos testigos: *uno que vive en la última casa de Northwestern Dr. y otro llamado Annabel que vive en Franklin Ave.*


````python
%%sql
SELECT *
FROM crime_scene_report
WHERE date = 20180115 AND city = 'SQL City';
````

**Paso 2: Encontrar a la Persona en la Última Casa de Northwestern Dr Y Anabel en Franklin Ave:**

Este resultado nos indica que **Morty Schapiro** vive en la última casa de Northwestern Dr, en el número 4919. Ademas, esta investigación muestra que **Annabel Miller** es quien vive en Franklin Ave.


````python
%%sql
SELECT *
FROM person
WHERE (address_street_name = 'Northwestern Dr' AND address_number = 4919)
   OR (address_street_name = 'Franklin Ave' AND name LIKE 'Annabel%');
````

**Paso 3: Revisar las Entrevistas de Morty Schapiro y Annabel Miller**

Para obtener las entrevistas de las personas encontradas en los pasos anteriores, necesitamos realizar una consulta basada en los id obtenidos.

**Resultados de las entrevistas:**

Morty Schapiro reportó haber escuchado un disparo y visto a un hombre huir
con una bolsa del gimnasio 'Get Fit Now', identificada como perteneciente a miembros oro cuyo número de membresía comienza con '48Z'. Además, observó al individuo subirse a un automóvil con una placa que incluía 'H42W'. Por otro lado, Annabel Miller identificó al asesino como alguien que había visto en el gimnasio la semana anterior al asesinato, el 9 de enero."

````python
%%sql
SELECT transcript
FROM interview
WHERE person_id IN (14887, 16371);
````

**Pista 4: Investigar a los miembros gold del gimnasio:**

Esta consulta busca en la base de datos de miembros del gimnasio "Get Fit Now" aquellos que tienen números de membresía que comienzan con "48Z" y que además tienen un estado de membresía "gold". Esta información nos ayuda a identificar a los miembros relevantes que podrían estar implicados en el crimen, basándonos en la evidencia de la escena del crimen y los testimonios recopilados.

````python
%%sql
SELECT *
FROM get_fit_now_member
WHERE id LIKE '48Z%' AND membership_status = 'gold'
````

**Paso 5: Investigación de Testimonios Clave:**

Este testimonio es fundamental ya que proporciona detalles significativos sobre una mujer desconocida que contrató al individuo en cuestión.
Según el testimonio:


*   Descripción Física y Vehicular: La mujer es descrita como alguien de aproximadamente 5'5" (65") o 5'7" (67") de estatura, con cabello rojo y conduce un Tesla Model S.

*   Patrones de Comportamiento: Además, se menciona que la mujer asistió al concierto SQL Symphony en tres ocasiones durante diciembre de 2017.

````python
%%sql
SELECT *
FROM interview
WHERE person_id IN (28819, 67318);
````

**Paso 6  En busca de la mujer rica:**

El objetivo de esta consulta fue identificar y listar todas las personas que poseen características físicas y vehiculares similares a las descritas por un testigo ocular clave. Estas personas potencialmente podrían estar relacionadas con el caso en cuestión, proporcionando pistas adicionales para nuestra investigación sobre el crimen.

````python
%%sql
SELECT p.name
FROM person p
JOIN drivers_license d ON p.license_id = d.id
JOIN facebook_event_checkin f ON p.id = f.person_id
WHERE f.event_name = 'SQL Symphony Concert'
  AND f.date >= 20171201 AND f.date <= 20171231
  AND d.hair_color = 'red'
  AND d.gender = 'female'
  AND d.car_model = 'Model S'
  AND d.height BETWEEN 65 AND 67
  AND d.car_make = 'Tesla'
  GROUP BY p.name
  HAVING COUNT(f.event_name) >= 3;
````
